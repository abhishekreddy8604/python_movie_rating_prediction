```python
#Importing libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression


# dataset Loading
df = pd.read_csv('movie_data.csv', encoding='ISO-8859-1')
```

```python

#Preprocessing: Filling missing values with the mean 
df.fillna(df.mean(numeric_only=True), inplace=True)

#Converting category data to numerical values using one-hot encoding
genres_dummies = df['genres'].str.get_dummies('|')
df = pd.concat([df, genres_dummies], axis=1)

#Drop non-numeric and unnecessary columns
df.drop(['genres', 'director_name', 'actor_2_name', 'actor_1_name', 'movie_title', 
         'actor_3_name', 'movie_imdb_link', 'language', 'country'], axis=1, inplace=True)
```

```python
#Creating Model


# Spliting the dataset into features X and target variable y
X = df.drop('imdb_score', axis=1)
Y = df['imdb_score']

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Create the Linear regression model 
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, Y_train)

#Make predictions on the test data
Y_pred = model.predict(X_test)


#Identify and handle outliers using a different approach
residuals = Y_test - Y_pred
outlier_mask = residuals.abs() > 2   # Identify outliers based on residuals

#Remove outliers from the testing set
X_test_filtered = X_test[~outlier_mask]
Y_test_filtered = Y_test[~outlier_mask]

#Make predictions on the filtered test set
Y_pred_filtered = model.predict(X_test_filtered)
Y_pred_filtered = np.clip(Y_pred_filtered, 0, 10)
```

```python
#Model Evaluation

#Calculating the Mean Squared Error MSE for the filtered model

mse_filtered = mean_squared_error(Y_test_filtered, Y_pred_filtered)

#Calculating the Mean Squared Error MSE for the model before filtering

mse = mean_squared_error(Y_test, Y_pred)

#Calculate the R-squared for the filtered model

r2_filtered = r2_score(Y_test_filtered, Y_pred_filtered)

#Calculate the R-squared for the model before filtering

r2 = r2_score(Y_test, Y_pred)

#Calculate the RMSE for the model before filtering

rmse_filtered = np.sqrt(mse_filtered)

#Calculate the R-squared for the filtered model

rmse = np.sqrt(mse)


print(f"Mean Squared Error (MSE) for Filtered Model: {mse_filtered}")
print(f"R-squared (R2) for Filtered Model: {r2_filtered}")
print(f"RMSE for Filtered Model: {rmse_filtered}")

print(f"Mean Squared Error (MSE) for Model before Filtering: {mse}")
print(f"R-squared (R2) for  Model before Filtering: {r2}")
print(f"RMSE for Filtered Model: {rmse}")


```

```python
#Visualization

#Actual vs Predicted IMDB Scores with Outliers

plt.figure(figsize=(10, 6))
sns.scatterplot(x=Y_test, y=Y_pred, alpha=0.5, color='blue')
sns.lineplot(x=[0, 10], y=[0, 10], color='red', linestyle='--', lw=2)
plt.title('Actual vs Predicted IMDb Scores (With Outliers)')
plt.xlabel('Actual scores')
plt.ylabel('Predicted scores')
plt.grid()
plt.show()


#Actual vs Predicted IMDB Scores with Outliers Highlated as per given mask

plt.figure(figsize=(10, 6))
sns.scatterplot(x=Y_test, y=Y_pred, alpha=0.5, label='Regular Points', color='blue')
sns.scatterplot(x=Y_test[outlier_mask], y=Y_pred[outlier_mask], alpha= 0.8, label='Outliers', color='red')
sns.lineplot(x=[0, 10], y=[0, 10], color='green', linestyle='--', lw=2)
plt.title('Actual vs Predicted IMDb Scores with Outliers Highlighted')
plt.xlabel('Actual IMDb Scores')
plt.ylabel('Predicted IMDb Scores')
plt.legend()
plt.grid()
plt.show()

#Actual vs Predicted IMDB Scores Withouit outliers

plt.figure(figsize=(10, 6))
sns.scatterplot(x=Y_test_filtered, y=Y_pred_filtered, alpha=0.5, color='green')
sns.lineplot(x=[0, 10], y=[0, 10], color='red', linestyle='--', lw=2)
plt.title('Actual vs Predicted IMDb Scores (Without Outliers)')
plt.xlabel('Actual scores')
plt.ylabel('Predicted scores')
plt.grid()
plt.show()



#residuals for the filtered test set
residuals_filtered = Y_test_filtered - Y_pred_filtered

#Histogram of Residuals for the filtered model
plt.figure(figsize=(10, 6))
sns.histplot(residuals_filtered, bins=20, kde=True, color='darkblue', edgecolor='black')
plt.title('Histogram of Residuals (Filtered Model)')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```
